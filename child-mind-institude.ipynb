```python
for column in fill_mode:
    mode_value = df[column].mode()[0]  # Mod değerini hesapla
    df[column] = df[column].fillna(mode_value)  # NaN değerlerini mod ile doldur
```

```python
for col in df.columns:
    if df[col].isnull().any():  
        mean_e = df[col].mean()
        df[col] = df[col].fillna(mean_e)
```

```python
le=LabelEncoder()
df["id"]=le.fit_transform(df["id"])
df["Basic_Demos-Enroll_Season"]=le.fit_transform(df["Basic_Demos-Enroll_Season"])
df["CGAS-Season"]=le.fit_transform(df["CGAS-Season"])
df["Physical-Season"]=le.fit_transform(df["Physical-Season"])
df["Fitness_Endurance-Season"]=le.fit_transform(df["Fitness_Endurance-Season"])
df["FGC-Season"]=le.fit_transform(df["FGC-Season"])
df["BIA-Season"]=le.fit_transform(df["BIA-Season"])
df["PAQ_A-Season"]=le.fit_transform(df["PAQ_A-Season"])
df["PAQ_C-Season"]=le.fit_transform(df["PAQ_C-Season"])
df["PCIAT-Season"]=le.fit_transform(df["PCIAT-Season"])
df["SDS-Season"]=le.fit_transform(df["SDS-Season"])
df["PreInt_EduHx-Season"]=le.fit_transform(df["PreInt_EduHx-Season"])
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
rf=RandomForestRegressor()
y=df["sii"]
x=df.drop("sii",axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=6,train_size=0.7)
```

```python
mode=lr.fit(x,y)
mode.score(x,y)
```

```python
model=lr.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
modell=dt.fit(x_train,y_train)
modell.score(x_test,y_test)
```

```python
modelll=rf.fit(x_train,y_train)
modelll.score(x_test,y_test)
```

```python

cv_scores = cross_val_score(model, x, y, cv=5) 
```

```python
print("Çapraz Doğrulama Skorları:", cv_scores)
print("Ortalama Skor:", cv_scores.mean())
```
